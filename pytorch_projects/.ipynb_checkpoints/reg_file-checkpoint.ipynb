```python
#!pip install torchmetrics

import torch
from torch import nn
from torch.utils.data import DataLoader, Dataset
#from torchvision import datasets
from sklearn import datasets
#from torchvision.transforms import ToTensor, Lambda, Compose
import torchmetrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

```python
a= datasets.load_wine()
class model(nn.Module):
  def __init__(self,in_features, out_features):
    super().__init__()
    self.in_features=in_features
    self.out_features=out_features
    self.Linear=nn.Linear(self.in_features, self.out_features)

  def forward(self, x):
    y_pred= self.Linear(x)
    return y_pred

class Data:
  def __init__(self, sample):
    self.sample = sample
    self.x = sample.data.astype(np.float32)
    self.y = sample.target.astype(np.float32)

  def __getitem__(self, index):
    return self.x[index], self.y[index]

  def __len__(self):
    return len(self.sample)
```

```python
data=Data(a)
train_data, test_data=train_test_split(data, test_size=.1, random_state=1, shuffle=True)
train_data=DataLoader(data, batch_size=3,shuffle=True)
b=iter(train_data)
print(next(b))
```

```python
epochs=100
model1=model(13,3)
criterion=nn.CrossEntropyLoss()
optim=torch.optim.Adam(model1.parameters(), lr=0.001)
for epoch in range(epochs):
  for i, (data1, label) in enumerate(train_data):
    label = label.long()
    y_pred=model1(data1)
    loss=criterion(y_pred, label)
    loss.backward()
    optim.step()
    optim.zero_grad()
    if epoch%10==0:
      print(loss.detach())
```
